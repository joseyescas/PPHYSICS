```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la gráfica
num_particles = 100  # Número de partículas
x_limits = (0, 10)   # Límites del eje X
y_limits = (0, 10)   # Límites del eje Y
particle_size = 50   # Tamaño de las partículas (constante para todas)

# Generar posiciones aleatorias para las partículas
x_positions = np.random.uniform(x_limits[0], x_limits[1], num_particles)
y_positions = np.random.uniform(y_limits[0], y_limits[1], num_particles)

# Crear la gráfica
plt.figure(figsize=(8, 8))
plt.scatter(x_positions, y_positions, s=particle_size, c='blue', alpha=0.7, edgecolors='black')

# Configurar límites y detalles
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Partículas en posiciones aleatorias')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la gráfica
num_lines = 10         # Número de líneas
particles_per_line = 10  # Número de partículas por línea
x_limits = (0, 10)      # Límites del eje X
y_limits = (0, 10)      # Límites del eje Y
particle_size = 50      # Tamaño de las partículas

# Generar posiciones para las partículas organizadas en líneas
x_positions = np.linspace(x_limits[0], x_limits[1], particles_per_line)  # Posiciones X uniformemente espaciadas
y_positions = np.linspace(y_limits[0], y_limits[1], num_lines)           # Posiciones Y uniformemente espaciadas

# Crear la gráfica
plt.figure(figsize=(8, 8))

# Dibujar las partículas línea por línea
for y in y_positions:
    plt.scatter(x_positions, [y] * particles_per_line, s=particle_size, c='blue', alpha=0.7, edgecolors='black')

# Configurar límites y detalles
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Partículas organizadas en líneas')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
